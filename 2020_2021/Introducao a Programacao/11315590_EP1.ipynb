```julia
function muro(N)
   soma = 0
    if  (N == 0)
        return 1
    else
        soma = 0

    if(N - 1) >= 0
            soma = soma +  muro(N - 1)
    end
        
    if (N - 2) >= 0
            soma = soma + 4 * muro(N - 2)
    end
        
    if (N - 3) >= 0
            soma = soma + 2 * muro(N - 3)
    end
    return soma
    end
end

muro(6)
```
