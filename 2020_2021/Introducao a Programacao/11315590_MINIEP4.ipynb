## Parte 2 - Comparando tempos

Vamos utilizar o macro `@elapsed` para calcular o tempo gasto em sua nova função `fibo2(n)` e comparar com a versão `fibo(n)` vista em sala. Rode o trecho de código abaixo, observe os tempos e responda: há diferença entre os tempos das duas versões? Qual o motivo por trás disso?

**Obs:** tente aumentar aumentar aos poucos o valor do argumento em `fibo(n)` e `fibo2(n)`. Tome cuidado, pois valores muito grandes podem demorar MUITO!

```julia
function fibo(n)
    if n <= 2
    else
        return fibo(n - 1) + fibo(n - 2)
    end
end
println("Tempo fibo(50): ", @elapsed fibo(50))
println("Tempo fibo2(50): ", @elapsed fibo(40))
```

Resposta: O tempo de resposta da função fibo2 é menor do que o tempo de resposta da função fibo. Isso acontece porque, na função fibo, o volume de operações feitas para a entrada do parametro n é maior do que as operações feitas individualmente para a entrada dos parâmetros n, a e b. Portanto, é normal que demore mais para processar as operações para a entrada do parâmetro a cada nova entrada feita na recursão e, consequentemente, que demore mais para a execução de cada versão.
