```python
import numpy as np
import matplotlib.pyplot as plt

def contraste_alongamento(img):
    min_val = np.percentile(img, 5)  # Obtm o valor mnimo de intensidade de 5% dos pixels
    max_val = np.percentile(img, 95)  # Obtm o valor mximo de intensidade de 95% dos pixels
    img_contrast = np.clip((img - min_val) / (max_val - min_val), 0, 1)  # Normaliza entre 0 e 1
    img_contrast *= 255  # Ajusta para a faixa de 0 a 255
    return img_contrast.astype(np.uint8)

img1_path = "/content/drive/MyDrive/img1.png"
img2_path = "/content/drive/MyDrive/img2.png"

img1 = plt.imread(img1_path)
img2 = plt.imread(img2_path)

# Converter para tons de cinza, se necessrio
img1 = np.mean(img1, axis=2)
img2 = np.mean(img2, axis=2)

# Aplicar contraste
img1_contrast = contraste_alongamento(img1)
img2_contrast = contraste_alongamento(img2)

# Subtrair a imagem com contraste aumentado da imagem original
img1_diff = img1_contrast - img1
img2_diff = img2_contrast - img2

# Plotar as imagens
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.imshow(img1, cmap='gray')
plt.title('Radiografia Trax Original')

plt.subplot(2, 3, 2)
plt.imshow(img1_contrast, cmap='gray')
plt.title('Radiografia Trax com Contraste Alongado')

plt.subplot(2, 3, 3)
plt.imshow(img1_diff, cmap='gray')
plt.title('Radiografia Trax Subtrao Contraste Alongado')

plt.subplot(2, 3, 4)
plt.imshow(img2, cmap='gray')
plt.title('Tomografia Crebro Original')

plt.subplot(2, 3, 5)
plt.imshow(img2_contrast, cmap='gray')
plt.title('Tomografia Crebro com Contraste Alongado')

plt.subplot(2, 3, 6)
plt.imshow(img2_diff, cmap='gray')
plt.title('Tomografia Crebro Subtrao Contraste Alongado')

plt.tight_layout()
plt.show()
```

podemos ver que a imagem do torax que originalmente possuia menos contrastre, ao aplicar a subtracao de alongamento de contraste, podemos visualizar caracteristicas que antes estavam pouco evidentes, pois a tcnica realca caracteristicas em imagens que possuem uma variacao de intensidade limitada. O mesmo nao acontece com a imagem do cerebro, pois a imagem original possui um alto contraste, uma alta variacao nos niveis de intensidade.
